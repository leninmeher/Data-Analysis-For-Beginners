```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.distplot(df['tip'],kde=False,bins=40)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker',palette='rainbow')
```

```python
sns.rugplot(df['total_bill'])
```

```python
sns.kdeplot(df['total_bill'])
```

```python
sns.kdeplot(df['tip'])
```

```python
sns.barplot(x='smoker',y='total_bill',data=df)
```

```python
sns.barplot(x='smoker',y='total_bill',data=df,estimator=np.mean)
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.boxplot(x='day',y='tip',hue='smoker',data=df,palette='rainbow')
```

```python
sns.violinplot(x='smoker',y='total_bill',hue='sex',data=df,palette='Set1')
```

```python
sns.violinplot(x='smoker',y='total_bill',hue='sex',data=df,split=True,palette='Set1')
```
